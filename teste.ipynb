```python
df = turma = pd.read_excel("..\..\Data\ALUNOS 1ªA.xlsx", skiprows=1)
alunos_lista = []

# Percorrendo cada linha do DataFrame
for _, row in df.iterrows():
    # Convertendo a data para o formato 'YYYY/MM/DD', se não for nula
    data_nascimento = None
    if pd.notnull(row.get("Data Nascimento")):
        data_nascimento = row["Data Nascimento"].strftime('%Y/%m/%d')
    
    # Tratando o campo 'codigoInep' para remover casas decimais
    codigo_inep = row.get("Código INEP", None)
    if pd.notnull(codigo_inep):
        codigo_inep = int(float(codigo_inep))  # Convertendo para inteiro

    # Montando o dicionário do aluno com os campos necessários
    aluno = {
        "nome": str(row.get("Nome do Aluno", None)),
        "filiacao1": str(row.get("Filiação 1", None)),
        "filiacao2": None,  # Garantindo que valores padrão também sejam string
        "nomeSocialAluno": str(row.get("Nome do Aluno", None)),
        "nomeAfetivoAluno": str(row.get("Nome do Aluno", None)),
        "dataNascimento": str(data_nascimento),
        "cpf": None,  # Valor padrão convertido para string
        "rg": row.get("rg", None),
        "codigoAluno": str(row.get("Código", None)),
        "codigoInep": str(codigo_inep),  # Removendo casas decimais
        "turma": None # Valor padrão convertido para string
    }
    
    # Adicionando o dicionário à lista
    alunos_lista.append(aluno)

alunos_lista
```
